```julia
using LinearAlgebra
using Plots
pyplot(size=(400,400))

n = 32
t = range(0, 2π, length=n+1)[1:n]
x = cos.(t)
y = sin.(t)
P = [x'; y']
scatter(x, y, leg=false)
xlims!(-2, 2)
ylims!(-2, 2)
```

```julia
function plot_aux(P, A)
    P2 = A * P
    scatter(P[1,:], P[2,:], leg=false)
    scatter!(P2[1,:], P2[2,:], c=:red)
    for i = 1:n
        plot!([P[1,i], P2[1,i]], [P[2,i], P2[2,i]], c=:green, l=:arrow)
    end
    xlims!(-6, 6)
    ylims!(-6, 6)
end
```

Matriz com autovalores reais distintos

```julia
V = [1.0 1.0; 1.0 3.0]
Λ = diagm(0 => [2.0; 4.0])
A = V * Λ * inv(V)
plot_aux(P, A)
plot!([0, V[1,1]], [0, V[2,1]], c=:gray, l=:arrow)
plot!([0, V[1,2]], [0, V[2,2]], c=:gray, l=:arrow)
```

```julia
V = [1.0 1.0; 1.0 3.0]
Λ = diagm(0 => [0.0; 4.0])
A = V * Λ * inv(V)
plot_aux(P, A)
plot!([0, V[1,1]], [0, V[2,1]], c=:gray, l=:arrow)
plot!([0, V[1,2]], [0, V[2,2]], c=:gray, l=:arrow)
```
