$$ \frac{1}{2\sqrt{a}} = \frac{d}{dx}\Big(\sqrt{x}\Big)|_{x = a} = \lim_{h \to 0} \frac{\sqrt{a + h} - \sqrt{a}}{h}
\approx \frac{\sqrt{a + h} - \sqrt{a}}{h}
= \frac{1}{\sqrt{a+h} + \sqrt{a}} $$

```julia
h = 10^(-15)
a = 1.0
1 / 2 / sqrt(a)
```

```julia
(sqrt(a + h) - sqrt(a)) / h
```
